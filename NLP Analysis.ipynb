```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datas = pd.read_csv('df_strength.csv')
```

```python
dataw = pd.read_csv('df_weakness.csv')
```

### TREK 1200 STRENGTH

```python
Trek1200_st_only = datas[(datas['category']=='Trek1200')]
```

#### sentiment

```python
sentiment_st = Trek1200_st_only[['Positive', 'Neutral', 'Negative']].mean()
sentiment_st.plot(kind='bar', title='Sentiment Scores', ylabel='Mean Sentiment Score', color = ['blue','gold','red'], figsize=(10, 7))
plt.savefig("Trek1200sentst.png")
```

### Emotion

```python
emotions_st = Trek1200_st_only[['trust', 'surprise', 'positive','joy', 'anticipation']].mean()
```

```python
emotions_st.plot(kind="pie", title='Emotion Scores', figsize=(10, 7),label="")
plt.savefig("Trek1200pie_st.png")
```

### KEYWORD

```python
Trek1200_Keyword = ' '.join([word for word in Trek1200_st_only['text_string_lem']])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=600, 
                     height=400, 
                     random_state=2, 
                     max_font_size=100).generate(Trek1200_Keyword)

plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off');
plt.savefig('Trek1200_Keyword_st.png')
```

```python
import yake
language = "en"
max_ngram_size = 1
deduplication_thresold = 0.9
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 10

kw_extractor = yake.KeywordExtractor(lan=language, 
                                     n=max_ngram_size, 
                                     dedupLim=deduplication_thresold, 
                                     dedupFunc=deduplication_algo, 
                                     windowsSize=windowSize, 
                                     top=numOfKeywords)
                                            
keywords = kw_extractor.extract_keywords(Trek1200_Keyword)

kwlist2=[]
for kw in keywords:
    print(kw)
    kwlist2.append(kw)
```

```python
kwtuple = pd.DataFrame(data = kwlist2, columns=['Keywword', 'Score'])
```

```python
kwtuple.to_csv('kwTrek1200.csv', index=False)
```
